```python
import pandas as pd
import klib as kl
```

```python
pcs_data = pd.read_csv('./validated_csv.csv', index_col=False)
project_type = pd.read_csv('./project_types.csv')
```

```python
pcs_data = pcs_data.iloc[:, 2:]
pcs_data.head(5)
```

```python
project_type = kl.clean_column_names(project_type)
project_type = project_type.iloc[:, 2:]
project_type.head(5)
```

```python
project_type = project_type[['project_category', 'construction_category', 'id']]
```

```python
merged = pd.merge(pcs_data, project_type, how='left', left_on='projectTypeId', right_on='id')
```

```python
merged = merged[['division', 'subdivision', 'item_code', 'description', 'quantity',
       'unit', ' labor_hourly_rate ', ' man_hrs_per_unit ',
       ' total_man_hours ', ' labor_total ', ' material_qty ',
       ' material_unit ', ' material ', ' material_total ', ' equipment_qty ',
       ' equipment_unit ', ' equipment ', ' equipment_total ', ' other_cost1 ',
       ' other_cost2 ', ' other_cost3 ', ' unit_cost ', ' unit_cost_total ',
       'subcontractor_qty', 'subcontractor_total', 'subcontractor',
       ' total_mat_lab_equip ', 'notes', 'project_id', 'project_type_id',
       'phase_id', 'content_id', 'file_id', 'file_name', 'source_file',
       'template_type', 'projectTypeId', 'projectId', 'phaseId', 'contentId',
       'fileId', 'source_file_name', 'project_category',
       'construction_category', 'id']]
```

```python
#encode source_file_name into projectId
merged['projectId'] = merged['source_file_name'].astype('category').cat.codes
```

```python
merged[' total_mat_lab_equip '] = merged[' total_mat_lab_equip '].str.replace(',', '').str.replace('$', '')
```
